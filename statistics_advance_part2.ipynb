1) **What is hypothesis testing in statistics?**

Hypothesis testing method is used to make inferences or draw conculsion about a population based on data.It helps you assess whether there's enough evidence to support a specific claim or hypothesis about a population paramenter.

**state the hypothesis**

You start by formulating two competing hypotheses:

**Null hypothesis (H₀):**

This is the hypothesis that there is no effect or difference, or it represents a position of "no change." It's the default assumption.

**Alternative hypothesis (H₁ or Ha)**:

This represents the claim you're testing for, which suggests that there is an effect or difference.

**Choose the Significance Level (α)**:

This is the threshold for deciding whether the evidence is strong enough to reject the null hypothesis. Common values are 0.05 (5%) or 0.01 (1%)

**Collect Data and Conduct the Test**:

You gather your sample data and perform a statistical test (like a t-test, chi-square test, ANOVA, etc.) to analyze the data.

**Calculate the p-value**:

The p-value tells you how likely it is to observe your sample data, or something more extreme, if the null hypothesis is true. A small p-value (typically less than α) suggests that the null hypothesis is unlikely, leading to its rejection

**Make a Decision: Based on the p-value:**

If p-value ≤ α, reject the null hypothesis (there’s evidence to support the alternative hypothesis).
If p-value > α, fail to reject the null hypothesis (there isn’t enough evidence to support the alternative hypothesis).

**Draw a Conclusion**:

Based on the decision, you either accept the alternative hypothesis or conclude that there's not enough evidence to reject the null hypothesis

2) **what is significance level in hypothesis testing, and why it is imporant?**

**What is the Significance Level?**
The significance level is a pre-defined probability that represents the likelihood of rejecting the null hypothesis when it is actually true.
Typically, α is set to 0.05, meaning you’re willing to accept a 5% chance of incorrectly rejecting the null hypothesis. Other common values for α include 0.01 (1%) or 0.10 (10%), depending on how stringent you want to be.

**Why is it Important?**

Control the Type I Error Rate:

A Type I error occurs when you reject the null hypothesis even though it’s actually true (a false positive).
The significance level α controls the probability of making this error. If you choose α = 0.05, for example, you're saying that you are willing to accept a 5% chance of making a Type I error.
Helps Make Objective Decisions:

The significance level sets a clear rule for decision-making. If the p-value from your test is smaller than α, you reject the null hypothesis. This helps ensure that decisions aren’t influenced by bias or subjectivity.
Influences the Power of the Test:

The significance level is closely related to the power of a statistical test, which is the probability of correctly rejecting the null hypothesis when it is false (avoiding a Type II error).
A smaller α (say, 0.01) will make it harder to reject the null hypothesis, but it reduces the likelihood of a Type I error. However, this can also reduce the power of the test, making it harder to detect a true effect if one exists.

**Example:**
Imagine you're testing a new drug, and your null hypothesis is "The drug has no effect." You set α = 0.05:

If your p-value is less than 0.05, you reject the null hypothesis and conclude that the drug has an effect (even though there could still be a small 5% chance that you’re wrong).
If the p-value is greater than 0.05, you fail to reject the null hypothesis, suggesting there’s not enough evidence to support the claim that the drug has an effect.

3) **what does p value represent in hypothesis testing?**

In hypothesis testing, the p-value is a measure that helps you assess the strength of the evidence against the null hypothesis. It tells you how likely it is to observe the sample data—or something more extreme—if the null hypothesis is true.

**What the p-value Represents:**

The p-value represents the probability of getting results at least as extreme as the ones observed in your sample, assuming that the null hypothesis is correct.

In simpler terms:

**A small p-value** (typically less than the significance level, α) indicates strong evidence against the null hypothesis, suggesting that you should reject the null hypothesis.
A large p-value suggests weak evidence against the null hypothesis, meaning you fail to reject it.
Interpreting the p-value:

**If p-value ≤ α (significance level):**

You reject the null hypothesis (H₀).
There’s enough evidence to support the alternative hypothesis (H₁).
For example, with α = 0.05 and p = 0.03, since 0.03 < 0.05, you reject the null hypothesis.

**If p-value > α (significance level):**

You fail to reject the null hypothesis.
There isn’t enough evidence to support the alternative hypothesis.
For example, with α = 0.05 and p = 0.12, since 0.12 > 0.05, you fail to reject the null hypothesis.


4) **How do you  interpret the p value in hypothesis testing?**

Interpreting the p-value in hypothesis testing is crucial for drawing conclusions about your data. Here's how you can interpret it step by step:

**1. Set the Significance Level (α)**

Before conducting the test, you decide on the significance level (α), which is the threshold that determines how strong the evidence must be to reject the null hypothesis. Common choices for α are 0.05, 0.01, or 0.10.

**2. Calculate the p-value**

Once you have your data and you've chosen a test (like a t-test or chi-square test), you calculate the p-value. This value tells you how likely it is to observe the data you collected (or something more extreme) if the null hypothesis (H₀) is true.

**3. Compare the p-value to α (the significance level)**

If p-value ≤ α (p-value is smaller than or equal to the significance level):

You reject the null hypothesis (H₀).
Interpretation: The data provides enough evidence to support the alternative hypothesis (H₁). In other words, the observed result is unlikely to have occurred by random chance, suggesting a significant effect or difference.
For example, if you set α = 0.05 and the p-value is 0.02, you reject H₀ and conclude that there is a statistically significant effect.
If p-value > α (p-value is greater than the significance level):

You fail to reject the null hypothesis.
Interpretation: The data does not provide enough evidence to support the alternative hypothesis. The observed result is consistent with the null hypothesis, meaning any observed difference or effect could be due to random chance.
For example, if α = 0.05 and the p-value is 0.10, you fail to reject H₀, suggesting there is no statistically significant effect.

5) **what are the type 1 and type 2 errors in hyothesis testing?**

**Type I Error (False Positive)**
A Type I error occurs when you incorrectly reject the null hypothesis (H₀) when it is actually true. In other words, you conclude that there is an effect or difference when, in reality, there isn't.

Example: Imagine you're testing a new drug to see if it has an effect on a disease. The null hypothesis (H₀) is that the drug has no effect. If you reject the null hypothesis and conclude that the drug does have an effect, even though it actually doesn't, you've made a Type I error.

Probability of a Type I error: The probability of committing a Type I error is denoted by α (the significance level), typically set to 0.05. This means you're willing to accept a 5% chance of making a Type I error.

Consequences: A Type I error leads you to believe there’s a significant effect or relationship when, in fact, there is none. This could lead to false conclusions, such as adopting ineffective treatments or making faulty business decisions.

**2. Type II Error (False Negative)**


A Type II error occurs when you fail to reject the null hypothesis (H₀) when it is actually false. In this case, you miss detecting a true effect or difference that exists.

Example: Using the same drug test, the null hypothesis (H₀) is that the drug has no effect. If you fail to reject the null hypothesis and conclude that the drug has no effect, even though it actually does, you've made a Type II error.

Probability of a Type II error: The probability of making a Type II error is denoted by β. The complement of β (1 − β) is called the power of the test, which is the probability of correctly rejecting the null hypothesis when it is false.

Consequences: A Type II error means failing to detect a real effect. This could lead to missed opportunities, such as not recognizing a beneficial treatment or ignoring important trends.


6) **What is diffrence between a one - tailed and two tailed test in hypothesis testing?**

In hypothesis testing, the difference between a one-tailed and a two-tailed test lies in the direction of the test and what you are testing for.

**One-tailed test:**

A one-tailed test tests for the possibility of an effect in only one direction.

It's used when you are interested in determining if the sample mean is significantly greater than or less than the population mean (but not both).

For example, if you're testing if a new drug is more effective than an existing one, you would use a one-tailed test to test if the new drug's mean is greater than the existing one.

Rejection region: All of the rejection region (where you reject the null hypothesis) is on one side of the distribution (either the left or the right, depending on the direction of the test).

**Two-tailed test:**

A two-tailed test tests for the possibility of an effect in both directions (either greater than or less than).

It's used when you want to check if the sample mean is significantly different from the population mean in either direction (it could be either greater than or less than).

For example, if you're testing whether a new drug has a different effect (could be more or less effective) compared to an existing one, you would use a two-tailed test.

**Rejection region**: The rejection region is split between both tails of the distribution. If the test statistic falls in either of the extreme ends (either side of the distribution), you reject the null hypothesis.


7) **What is z test and when is it used in hypothesis testing?**

A z-test is a statistical test used to determine whether there is a significant difference between the means of two groups or whether a single group mean differs from a known value, based on a sample from the population.

When is a z-test used in hypothesis testing?
A z-test is commonly used in the following scenarios:

**Single Sample Mean (One-sample z-test):**

When you're comparing the sample mean to a known population mean.
Example: Is the average weight of apples in a sample different from the known population average of 150 grams?

**Two Sample Means (Two-sample z-test):**

When you're comparing the means of two independent samples (both samples come from normal populations with known standard deviations).
Example: Do two different groups of students have the same average test scores?
Proportions (Z-test for Proportions):

When you're comparing proportions (like success rates) between groups.
Example: Does the proportion of people who favor a new product differ between two regions?

13) **what is confidence interval,and how is it used to interpret statistical results?**


A confidence interval (CI) is a range of values that is used to estimate an unknown population parameter (like a mean or proportion) based on sample data. It provides a range of plausible values for that parameter, along with a degree of confidence that the true parameter lies within that range.

For example, a 95% confidence interval means that if you were to take 100 different samples and calculate the confidence interval for each one, approximately 95 of those intervals would contain the true population parameter.

**Key Points:**

**Range of Values**: The interval gives you a range (e.g., 50 to 60) within which the true parameter is likely to fall.

**Confidence Level**: The confidence level (e.g., 95% or 99%) tells you how confident you can be that the interval contains the true population parameter. A higher confidence level means a wider interval.

**Not a Guarantee**: A CI doesn't guarantee that the true value is in the range, just that, based on your sample data, you're X% confident that the true value lies in this range.


14) **What is margin of error ,and how does it affect the confidence interval?**


The margin of error (MOE) is a measure of the uncertainty or precision of an estimate from sample data. It represents the range within which the true population parameter (like the population mean or proportion) is expected to fall, considering the sample size and variability in the data.

**Relationship with Confidence Interval:**

The margin of error directly impacts the width of the confidence interval (CI). The CI is typically expressed as:

CI = Estimate ± Margin of Error

In other words, the margin of error defines how far the estimate might reasonably be from the true population parameter.

How Margin of Error Affects Confidence Interval:
Larger Margin of Error: The confidence interval will be wider. This means the estimate is less precise, and there’s greater uncertainty about where the true parameter lies. A wider interval reflects less confidence in the precision of the estimate.

Smaller Margin of Error: The confidence interval will be narrower, indicating more precision and a higher level of certainty about where the true parameter is likely to be.

1)**Write a python program to perform a z - test for comparing a sample mean to a known population mean and interpret the result.**

```python
import math
from scipy.stats import norm
def Z_test(sample_mean, population_mean,population_std, sample_size,alpha=0.05):
    z_score = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))
    z_critical = norm.ppf(1 - alpha/2)
    result =  abs(z_score) > z_critical
    if result:
        print("Reject the null hypothesis")
    else:
        print("we are fail to reject null hypothesis")
    return z_score, z_critical, result
```

2) **simulate random data to perform hypothesis testing and calculate the corresponding p-value using python.**

```python
import numpy as np
from scipy import stats
np.random.seed(0)
sample_size = 20
population_mean = 500
population_std = 15
sample = np.random.normal(loc=population_mean,scale=population_std,size=sample_size)
sample_mean = np.mean(sample)
sample_std = np.std(sample,ddof = 1)
t_statistic, p_value = stats.ttest_1samp(sample, population_mean)
alpha = 0.05
print(f"sample_mean:{sample_mean:.2f}")
print(f"sample_std:{sample_std:.2f}")
print(f"t_statistic:{t_statistic:.2f}")
print(f"p_value:{p_value:.4f}")
if p_value < alpha:
    result = "Reject the null hypothesis"
else:
    result = "fail to reject the null hypothesis"
print(result)


```

3) **Implement a one-sample Z-test using python and visualize the decision region on a plot**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
def hypothesis_testing_with_errors(population_mean,population_std,sample_size,alpha,beta):
    sample_null = np.random.normal(loc = population_mean,scale = population_std,size = sample_size)
    sample_mean_null = np.mean(sample_null)
    sample_alternative = np.random.normal(loc = population_mean+5,scale = population_std,size = sample_size)
    sample_mean_alternative = np.mean(sample_alternative)
    z_score_null = (sample_mean_null - population_mean)/(population_std/np.sqrt(sample_size))
    z_score_alternative = (sample_mean_alternative - population_mean)/(population_std/np.sqrt(sample_size))
    z_critical_null = norm.ppf(1 - alpha)
    z_critical_alternative = norm.ppf(1-beta)
    type1_error_area = z_score_null>z_critical_null
    type2_error_area =  abs(z_score_alternative)<z_critical_alternative
    x = np.linspace(-4,4,100)
    y = norm.pdf(x,0,1)
    plt.figure(figsize = (8,6))
    plt.plot(x,y,0,label = "standard normal distribution",color = "blue")
    plt.fill_between(x,y,0,where = (x < -z_critical)| (x > z_critical),color = "red",alpha = 0.05,label = "rejection region")
    plt.axvline(z_score_null,color = "green",linestyle = "--",label = f"z_statistics (HO) = {z_score_null}")
    plt.axvline(z_score_alternative,color = "red",linestyle = "--",label = f"z_statistics (H1) = {z_score_alternative}")
    plt.title("Hypothesis Testing: Type I and Type II Errors")
    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.legend()
    print(f"Sample Mean under H0: {sample_mean_null:.2f}")
    print(f"Z-Statistic under H0: {z_score_null:.2f}")
    print(f"Sample Mean under H1: {sample_mean_alternative:.2f}")
    print(f"Z-Statistic under H1: {z_score_alternative:.2f}")
    print(f"Type I Error (Reject H0 when H0 is true): {'Yes' if type1_error_area else 'No'}")
    print(f"Type II Error (Fail to reject H0 when H1 is true): {'Yes' if type2_error_area else 'No'}")
    plt.show()


```

9) **Write a python function to calculate the confidence interval for a sample mean  and explain its significance.**

```python
import scipy.stats as stats
import numpy as np
def calculate_confidence_interval(sample,confidence_level = 0.95):
  sample_mean = np.mean(sample)
  sample_std = np.std(sample,ddof = 1)
  sample_size = len(sample)
  if sample_size > 30:
    critical_value = stats.norm.ppf(1-(1-confidence_level)/2)
  else:
    critical_value = stats.t.ppf(1-(1-confidence_level)/2,df = sample_size-1)
  margin_error = critical_value * (sample_std / np.sqrt(sample_size))
  lower_bound = sample_mean - margin_error
  upper_bound = sample_mean + margin_error
  return (lower_bound,upper_bound)
sample_data = [12, 15, 18, 14, 19, 22, 16, 21, 20, 18]
ci = calculate_confidence_interval(sample_data)
print(f"The 95% confidence interval for the sample mean is: {ci}")
```

```python
import numpy as np
def calculate_expected_frequencies(observed):
  row_totals = observed.sum(axis = 1)
  col_totals = observed.sum(axis = 0)
  grand_total = observed.sum()
  expected = np.outer(row_totals,col_totals)/grand_total
  return expected
observed = np.array([[20,30],[40,50]])
expected_frequencies = calculate_expected_frequencies(observed)
print("observed frequencies:\n ",observed)
print("\nexpected frequencies:\n",expected_frequencies)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=50, scale=20, size=100)
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)
F_statistic = var1 / var2
df1 = len(data1) - 1
df2 = len(data2) - 1
p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))
alpha = 0.05
print(f"Sample Variance 1: {var1}")
print(f"Sample Variance 2: {var2}")
print(f"F-statistic: {F_statistic}")
print(f"P-value: {p_value}")
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
plt.figure(figsize = (8,6))
plt.subplot(1,2,1)
plt.hist(data1,alpha = 0.7 ,bins = 20,label = "dataset1",color = "blue")
plt.hist(data2,alpha = 0.7 ,bins = 20,label = "dataset2",color = "orange")
plt.legend()
plt.subplot(1,2,2)
plt.boxplot([data1, data2], labels=['Dataset 1', 'Dataset 2'],patch_artist=True)
plt.title("boxplots of Dataset 1 and Dataset 2")
plt.show()
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
np.random.seed(42)
observed = np.random.choice([1, 2, 3, 4, 5, 6], size=1000, p=[1/6]*6)
observed_counts = np.array([np.sum(observed == i) for i in range(1, 7)])
expected_counts = np.array([1000 * (1/6)] * 6)
chi2_stat, p_value = stats.chisquare(observed_counts, expected_counts)
alpha = 0.05
print(f"Observed Frequencies: {observed_counts}")
print(f"Expected Frequencies: {expected_counts}")
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
if p_value < alpha:
    print("Reject the null hypothesis: The observed frequencies are significantly different from the expected frequencies.")
else:
    print("Fail to reject the null hypothesis: The observed frequencies are not significantly different from the expected frequencies.")
labels = [1, 2, 3, 4, 5, 6]
plt.figure(figsize=(10, 6))
bar_width = 0.35
x = np.arange(len(labels))
plt.bar(x - bar_width/2, observed_counts, bar_width, label='Observed', color='blue')
plt.bar(x + bar_width/2, expected_counts, bar_width, label='Expected', color='orange')
plt.xlabel('Die Faces')
plt.ylabel('Frequency')
plt.title('Chi-Square Goodness of Fit: Observed vs Expected Frequencies')
plt.xticks(x, labels)
plt.legend()
plt.show()
```
